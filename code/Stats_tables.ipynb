```python
import numpy as np
import pandas as pd
from statsmodels.stats.multitest import multipletests
```

```python
# define path
path = '/Users/laurituominen/Documents/Research/Reettis/neuromaps/'

# get data for plotting 
corrs_turku = dict(np.load(path + 'data/corrs_turku.npz'))

corrs_enigma= dict(np.load(path + 'data/corrs_enigma.npz'))
```

```python
sources = ['kaller2017_sch23390','jaworska2020_fallypride', 'beliveau2017_dasb', 'savli2012_way100635',
             'beliveau2017_az10419369', 'beliveau2017_cimbi36', 'beliveau2017_sb207145', 'radnakrishnan2018_gsk215083',
             'tuominen_feobv', 'hillmer2016_flubatine', 'naganawa2020_lsn3172176', 'margulies2016_fcgradient01', 
             'hcps1200_megalpha','hcps1200_megbeta', 'hcps1200_megdelta', 'hcps1200_meggamma1', 'hcps1200_meggamma2',
             'hcps1200_megtheta', 'hcps1200_megtimescale','finnema2016_ucbj', 'hcps1200_thickness', 
             'hcps1200_myelinmap', 'dukart2018_flumazenil', 'dubois2015_abp688', 'laurikainen2018_fmpepd2',
             'kantonen2020_carfentanil','gallezot2017_gsk189254', 'raichle_cbf', 'raichle_cbv', 'raichle_cmr02', 
             'raichle_cmruglu']
```

```python
df_turku = pd.DataFrame(corrs_turku).transpose()
df_turku.columns = ['rho', 'pspin']
df_turku = df_turku[df_turku.index.isin(sources)]
_, fdr_corrected_pvalues, _, _ = multipletests(df_turku['pspin'], alpha=0.05, method='fdr_bh')
df_turku['fdr_corrected_p_value'] = fdr_corrected_pvalues

display(df_turku)

df_turku.to_csv(path + '/tables/correlations_in_discovery_sample.csv')
```

```python
index_fdr = df_turku.index[df_turku['fdr_corrected_p_value']< 0.05] 

df_enigma = pd.DataFrame(corrs_enigma).transpose()
df_enigma.columns = ['rho', 'pspin']
df_enigma = df_enigma[df_enigma.index.isin(index_fdr)]
_, fdr_corrected_pvalues, _, _ = multipletests(df_enigma['pspin'], alpha=0.05, method='fdr_bh')

df_enigma['fdr_corrected_p_value'] = fdr_corrected_pvalues
display(df_enigma)

df_enigma.to_csv(path + '/tables/correlations_in_replication_sample.csv')
```
