```python
from surfplot import Plot
from brainspace.datasets import load_parcellation
from brainspace.mesh.mesh_io import read_surface
from neuromaps.datasets import fetch_fsaverage
import os
import numpy as np
from neuromaps.images import construct_shape_gii, load_data
```

```python
# path
base_path = os.path.dirname(os.getcwd())

# load data 
img_L = load_data(os.path.join(base_path, 'data', 'lh.sig.nii'))
img_R = load_data(os.path.join(base_path, 'data', 'rh.sig.nii'))

# threshold
img_L[np.where(img_L > -1.3)] = 0
img_R[np.where(img_R > -1.3)] = 0

turku_map = {'left':img_L, 'right':img_R}

# load template
surface = fetch_fsaverage(density='164k')
lh, rh = surface['inflated']

# plot
p = Plot(lh, rh, views=['lateral','medial'], zoom=1.5)
p.add_layer(turku_map, cbar=True, cmap='inferno',  color_range=(-5,-1.3),zero_transparent=True)
pic = p.build()
```
