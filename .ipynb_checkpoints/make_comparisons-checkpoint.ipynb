```python

from neuromaps.images import load_gifti
from neuromaps import transforms
from neuromaps.stats import compare_images
from neuromaps import nulls
from neuromaps.resampling import resample_images
from neuromaps.datasets import fetch_annotation
from neuromaps.plotting import plot_surf_template
import matplotlib.pyplot as plt
```

```python

# get thickness results  
d = '/Users/laurituominen/Documents/Research/Reettis/Laurille_131022/plots/combine_lhrh_gifti/'
sig_lh = load_gifti(d + 'lh.sig.mgh.gii')
sig_rh = load_gifti(d + 'rh.sig.mgh.gii')

sig = (sig_lh, sig_rh)
```

```python
# plot sig
plt.ion()
plot_surf_template(sig, 'fsaverage', '164k')
```

```python

# transform gamma from fsaverage to fsLR
sig_fslr = transforms.fsaverage_to_fslr(sig, '164k')

# get annotations 
fsLR_annotations = fetch_annotation(source=['hcps1200',  
                                         'margulies2016'])

# compare gamma to annotations using rotated nulls 
for annot in fsLR_annotations.keys():
    sig_res, annot_res = resample_images(src=sig_fslr, trg=fsLR_annotations[annot],
                                  src_space='fsLR', trg_space='fsLR',
                                  method='linear', resampling='downsample_only')
 
    rotated = nulls.alexander_bloch(annot_res, atlas='fsLR', density=annot[3],
                                    n_perm=1000, seed=1234)

    corr, pval = compare_images(annot_res, sig_res, nulls=rotated)
    print(annot)
    print(f'r = {corr:.3f}, p = {pval:.3f}')
```

```python
# do raichle

# get annotations 
raichle_annotations = fetch_annotation(source=['raichle'])

# compare gamma to annotations using rotated nulls 
for annot in raichle_annotations.keys():
    sig_res, annot_res = resample_images(src=sig_fslr, trg=raichle_annotations[annot],
                                  src_space='fsLR', trg_space='fsLR',
                                  method='linear', resampling='downsample_only')
 
    rotated = nulls.alexander_bloch(annot_res, atlas='fsLR', density=annot[3],
                                    n_perm=1000, seed=1234)

    corr, pval = compare_images( annot_res, sig_res, nulls=rotated)
    print(annot)
    print(f'r = {corr:.3f}, p = {pval:.3f}')
```
